```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country
```

```python
df.columns
```

```python
# How can we merge both files.
df = pd.merge(df, df_country, on="Country Code", how="left")
df.head()

```

```python
df.head()
```

```python
# how can we rename any column
df.rename(columns={"Country_x":"Country"}, inplace=True)
```

```python
df.head(2)
```

```python
# To check data types
df.dtypes
```

```python
country_name = df.Country.value_counts().index
```

```python
country_value = df.Country.value_counts().values
```

```python
# creating pie chart

plt.figure(figsize=(10,10))
plt.pie(country_value[:3], labels=country_name[:3], autopct="%1.0f%%", shadow=True, startangle=90)
plt.show()
```

```python
# using groupby function for co_relation between columns
ratings = df.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})

```

# Conclusions from uper groupby functions
1. see maximum people not giving ratings.
2. most of the people who are giving ratings are give "Average" ratings and so on 

```python
# using seaborn for plotings ...we plot bar plot
plt.figure(figsize=(5,5))
sns.barplot(x="Aggregate rating", y="Rating color", data=ratings)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x="Aggregate rating", y="Rating count", data=ratings)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x="Aggregate rating", y="Rating count", hue="Rating color", data=ratings)
plt.show()
```

```python
# count plot with the help of categorical variables
plt.figure(figsize=(10,10))
sns.countplot(x="Rating text", data=ratings)
plt.show()
```

```python
# count plot with the help of categorical variables
plt.figure(figsize=(10,10))
sns.countplot(x="Rating color", data=ratings)
plt.show()
```

```python
# Find the countries name that has given 0 ratings
df[df["Aggregate rating"]==0]["Country"].value_counts().reset_index()

```

```python
# find out the currency which is used as a payment in different countries
df["Currency"].value_counts().reset_index()
```

```python
# find out currency in a better way
df.groupby(["Country","Currency"]).size().reset_index().rename(columns={0:"Count"})
```

```python
# find which countries do have online deliveries with name column
df.groupby(["Country","Has Online delivery"]).size().reset_index().rename(columns={0:"Count"})
```

```python
# creating a pie chaart for citiea distribution

plt.pie(df["City"].value_counts().values[:5], labels=df["City"].value_counts().index[:5], autopct="%1.0f%%")
plt.show()
```
