### Feature Engineering process starts

```python
final_df.head()
```

```python
# Spliting date column into day , month and year
final_df["Day"] = pd.to_datetime(final_df.Date_of_Journey, format="%d/%m/%Y").dt.day
final_df["Month"] = pd.to_datetime(final_df.Date_of_Journey, format="%d/%m/%Y").dt.month
final_df["Year"] = pd.to_datetime(final_df.Date_of_Journey, format="%d/%m/%Y").dt.year
final_df.head()
```

```python
final_df.info()
```

```python
# dropping column
final_df.drop(["Date_of_Journey"], axis = 1, inplace = True)
final_df.head(1)
```

```python
# Feature engineering on Arrival_Time column with the help of split function 
final_df["Arrival_Time"] = final_df["Arrival_Time"].str.split(" ").str[0]
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
# Feature engineering on Arrival_Time column with the help of split function also convert it into int type
final_df["Arrival_hour"] = final_df["Arrival_Time"] .str.split(":").str[0].astype(int)
final_df["Arrival_min"] = final_df["Arrival_Time"] .str.split(":").str[1].astype(int)
final_df.head()
```

```python
# dropping column
final_df.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
final_df.head(1)
```

```python
# feature engineering on Dep_Time 
final_df["Dep_hour"] = final_df["Dep_Time"] .str.split(":").str[0].astype(int)
final_df["Dep_min"] = final_df["Dep_Time"] .str.split(":").str[1].astype(int)
final_df.head()
```

```python
# Dropping column
final_df.drop(["Dep_Time"], axis = 1, inplace = True)
final_df.head(1)
```

```python
final_df.info()
```

```python
#looking for nan value in Total_Stops column exact position
final_df[final_df["Total_Stops"].isnull()]
```

```python
# Handling Total_Stops column with the help map function and also convert in int type

final_df["Total_Stops"] = final_df["Total_Stops"].map({"non-stop":0, "1 stop":1, "2 stops":2, "3 stops":3, "4 stops":4, "NaN":1})
```

```python
final_df.head()
```

```python
# Dropping column
final_df.drop(["Route"], axis = 1, inplace = True)
final_df.head(1)
```

```python
final_df.head()
```

```python
# Handle Duratin column in a better way 
duration = list(final_df["Duration"])
for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour
duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
final_df["Duration_hours"] = duration_hours
final_df["Duration_mins"] = duration_mins
final_df.head()
```

```python
# Dropping column
final_df.drop(["Duration"], axis = 1, inplace = True)
final_df.head(1)
```

```python
# Using label encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
final_df["Airline"] = encoder.fit_transform(final_df["Airline"])
final_df.head()
# Using One Hot Encoding
Airline = final_df[["Airline"]]
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
# Using Label Encoding
final_df["Source"] = encoder.fit_transform(final_df["Source"])
final_df.head()
# Using One Hot Encoding
Source = final_df[["Source"]]
Source = pd.get_dummies(Source, drop_first= True)
Source.head()
# Using Label Encoding
final_df["Destination"] = encoder.fit_transform(final_df["Destination"])
final_df.head()
# Using One Hot Encoding
Destination = final_df[["Destination"]]
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
# Using Label Encoding
final_df["Additional_Info"] = encoder.fit_transform(final_df["Additional_Info"])
final_df.head()
# Using One Hot Encoding
Additional_Info = final_df[["Additional_Info"]]
Additional_Info = pd.get_dummies(Additional_Info, drop_first = True)
Additional_Info.head()
# Using Label Encoding
final_df["Total_Stops"] = encoder.fit_transform(final_df["Total_Stops"])
final_df.head()
# Concatenate dataframe --> final_df + Airline + Source + Destination
# final_df = pd.concat([final_df, Airline, Source, Destination, Additional_Info], axis = 1)
# final_df.head()
# Dropping column
# final_df.drop(["Airline", "Source", "Destination", "Additional_Info"], axis = 1, inplace = True)
```
